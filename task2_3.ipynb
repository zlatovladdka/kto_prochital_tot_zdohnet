```python
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib.patches import Ellipse
%matplotlib inline
```

```python
a = np.array([[3,0],[0,-2]])
u, s, vh = np.linalg.svd(a)
```

```python
alpha1 = np.angle(u[0,0] + 1j*u[1,0])

pic = plt.subplot(aspect = 'equal')
ellipse = Ellipse((0, 0), 2*np.linalg.norm([*u[:,0]*s[0]]), 2*np.linalg.norm([*u[:,1]*s[1]]), alpha1*180, facecolor='none', edgecolor='violet')
pic.add_artist(ellipse)
plt.arrow(0,0,*u[:,0]*s[0],color='blue')
plt.arrow(0,0,*u[:,1]*s[1],color='red')
plt.ylim(-4,4)
plt.xlim(-4,4)

plt.show()
```

```python
alpha2 = np.angle(vh[0,0] + 1j*vh[0,1])

pic = plt.subplot(aspect = 'equal')
ellipse = Ellipse((0, 0), 2*np.linalg.norm([*vh[0,:]]), 2*np.linalg.norm([*vh[1,:]]), alpha2*180, facecolor='none', edgecolor='violet')
pic.add_artist(ellipse)
plt.arrow(0,0,*vh[0,:],color='blue')
plt.arrow(0,0,*vh[1,:],color='red')
plt.ylim(-2,2)
plt.xlim(-2,2)
```

```python
fig = plt.figure(figsize=plt.figaspect(1))  # Square figure
ax = fig.add_subplot(111, projection='3d')

coefs = (np.linalg.norm([*u[:,0]*s[0]]), np.linalg.norm([*u[:,1]*s[1]]), np.linalg.norm([*u[:,2]*s[len(s)-1]]))
rx, ry, rz = 1/np.sqrt(coefs)


u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)


x = rx * np.outer(np.cos(u), np.sin(v))
y = ry * np.outer(np.sin(u), np.sin(v))
z = rz * np.outer(np.ones_like(u), np.cos(v))


ax.plot_surface(x, y, z,  rstride=4, cstride=4, color='b')


max_radius = max(rx, ry, rz)
for axis in 'xyz':
    getattr(ax, 'set_{}lim'.format(axis))((-max_radius, max_radius))

plt.show()
```

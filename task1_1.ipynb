```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
import pandas as pd
```

прямая рекурсия для $\alpha = 0.1$

```python
a = 0.1
I1 = np.zeros(50)
I1[0] = np.log(abs(1+a)/abs(a))
for i in range(1,len(I1)):
    I1[i] = 1/i - a*I1[i-1]
I1[25]
```

обратная рекурсия для того же $\alpha$

```python
N = 50
I2 = np.zeros(N)
for i in range(2, N):
    I2[N-i] = (1/(N-i+1) - I2[N-i+1])/a
I2[25]
```

прямая рекурсия для $\alpha = 10$

```python
a = 10
I3 = np.zeros(50)
I3[0] = np.log(abs(1+a)/abs(a))
for i in range(1,50):
    I3[i] = 1/i - a*I3[i-1]
I3[25]
```

```python
N = 50
I4 = np.zeros(N)
for i in range(2, N):
    I4[N-i] = (1/(N-i+1) - I4[N-i+1])/a
I4[25]
```

```python
x = sympy.Symbol('x')
N = 50
a = 0.1
exact01 = [float(sympy.integrate(x**n/(1 + a), (x, 0, 1))) for n in range(N)]
a = 10
exact10 = [float(sympy.integrate(x**n/(1 + a), (x, 0, 1))) for n in range(N)]
```

```python
diff1 = []
for value, exact_value in zip(I1, exact01):
    diff1.append(abs(value-exact_value))
```

```python
diff2 = []
for value, exact_value in zip(I2, exact01):
    diff2.append(abs(value-exact_value))
```

```python
diff3 = []
for value, exact_value in zip(I3, exact10):
    diff3.append(abs(value-exact_value))
```

```python
diff4 = []
for value, exact_value in zip(I4, exact01):
    diff4.append(abs(value-exact_value))
```

```python
df = pd.DataFrame({'0.1 upwards':diff1,'0.1 downwards':diff2, '10 upwards':diff3,'10 downwards':diff4})
df
```

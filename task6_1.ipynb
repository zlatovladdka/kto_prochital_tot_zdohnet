```python
import numpy as np
import matplotlib.pyplot as plt
```

$x^2 = 2$

```python
n = 8
x = np.zeros(n)
x[0] = 1
```

```python
for i in range(1, n):
    x[i] = (x[i-1] + 2/x[i-1])/2
```

```python
y = np.sqrt(2)
```

```python
plt.plot((x - y)**2)
plt.yscale('log')
```

получаем квадратичную сходимость

$1 + \cos{x} = 0$

```python
n = 30
x = np.zeros(n)
x[0] = 1
```

```python
for i in range(1, n):
    x[i] = x[i-1] + (np.cos(x[i-1])+1)/np.sin(x[i-1])
```

```python
y = np.pi
```

```python
plt.plot((x - y)**2)
plt.yscale('log')
```

в этом случае сходимость линейная

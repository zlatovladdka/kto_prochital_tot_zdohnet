```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with np.load('data.npz') as data:
    A, C = data['A'], data['C']
```

```python
def mat2vec(A):
    h, w = A.shape
    a = np.zeros(h*w, dtype=A.dtype)
    A = np.flipud(A) 
    for i, row in enumerate(A):
        a[i*w:i*w+w] = row
    return a
```

```python
def vec2mat(a, shape):
    h, w = shape
    A = np.zeros(shape, dtype=a.dtype)
    for i in range(h):
        A[i, :] = a[i*w:i*w+w]
    return np.flipud(A)
```

```python
plt.imshow(vec2mat(mat2vec(A), (25,60)))
```

```python
B0 = np.zeros((16,51))
B0[2:-2,1] = 1; B0[2,2:6] = 1
B0[2:7,6] = 1; B0[7:-2, 7] = 1
B0[7, 2:7] = 1; B0[-3, 2:7] = 1
B0[2:-2, 10] = 1; B0[2:-2, 14] = 1
B0[2:-2, 18] = 1; B0[-3, 10:19] = 1
B0[2:-2, 26] = 1; B0[2, 21:26] = 1
B0[7, 21:26] = 1; B0[-3, 21:26] = 1

B0[2:-2, 36] = 1; B0[2, 31:36] = 1
B0[7, 31:36] = 1; B0[-3, 31:36] = 1

B0[2:-2, 46] = 1; B0[2, 41:46] = 1
B0[7, 41:46] = 1; B0[-3, 41:46] = 1
plt.imshow(B0)
```

```python
b0 = mat2vec(B0)
b = C @ b0
B = vec2mat(b, (25, 60))
```

```python
plt.imshow(B)
```

```python
b0_recover = np.linalg.pinv(C) @ b
B0_recover = vec2mat(b0_recover, (16, 51))
```

```python
plt.imshow(B0_recover)
```

```python
a = mat2vec(A)
a0_recover = np.linalg.pinv(C) @ a
A0_recover = vec2mat(a0_recover, (16, 51))
```

```python
plt.imshow(A0_recover)
```

```python
def pinv_new(M, n):
    u, s, vh = np.linalg.svd(M)
    s[n:] = np.zeros(len(s) - n)
    s[:n] = s[:n] ** (-1)
    sigma = np.zeros((16*51, 25*60))
    for i in range(n):
        sigma[i,i] = s[i]
    return vh.T @ sigma @ u.T
```

```python
a = mat2vec(A)
a0_recover = pinv_new(C, 100) @ a
A0_recover = vec2mat(a0_recover, (16, 51))
plt.imshow(A0_recover)
```

```python
a = mat2vec(A)
a0_recover = pinv_new(C, 200) @ a
A0_recover = vec2mat(a0_recover, (16, 51))
plt.imshow(A0_recover)
```
